```python
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

from magrathea.maps.rendering_engine import generate_heightmap

# Configure matplotlib to display inline
%matplotlib inline
```

```python
# Parameters
size = 256
octaves = 4
seed = 42
island_density = 0.0

# Generate Map
heightmap = generate_heightmap(
    size=size, octaves=octaves, seed=seed, island_density=island_density
)

# Visualize

colors = [
    (0.0, "#1f4fff"),  # blue (deep water)
    (0.4, "#1f4fff"),  # blue (deep water)
    (0.5, "#f5e663"),  # yellow (sand)
    (0.55, "#4caf50"),  # green (grass)
    (0.8, "#0b3d0b"),  # dark green (dense forest)
    (1.0, "#0b3d0b"),  # dark green (dense forest)
]

sea_sand_grass = LinearSegmentedColormap.from_list("sea_sand_grass", colors, N=256)


plt.figure(figsize=(10, 10))
plt.imshow(heightmap, cmap=sea_sand_grass, origin="upper", vmin=0, vmax=1)
plt.colorbar(label="Elevation")
plt.title(f"Map Seed: {seed}")
plt.axis("off")
plt.show()
```
